Main loop

```python
import os
import pandas as pd

# Directory containing the CSV files
directory = './data/sentiment'

# List to store the results
results = []

# Process each CSV file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        # Read the CSV file
        file_path = os.path.join(directory, filename)
        df = pd.read_csv(file_path)

        # Calculate the average sentiment score
        average_sentiment = df['sentiment score'].mean()

        # Extract the first video ID (assuming all comments in a file are for the same video)
        video_id = df['video ID'].iloc[0]

        # Extract the first song ID (assuming all comments in a file are for the same song)
        song_id = df['song_id'].iloc[0]

        # Append the results
        results.append({'video ID': video_id, 'song_id': song_id, 'sentiment': average_sentiment})

# Create a new DataFrame with the results
result_df = pd.DataFrame(results)

result_df.head()
```
